```python
import json
import matplotlib.pyplot as plt
```

```python
with open('./data.json', 'r') as file:
    data = json.load(file)
```

```python
data
```

```python
for app, tests in data.items():
    for test_source, test_data in tests.items():
        for test_type, values in test_data.items():
            print(f"{app} : {test_source} : {test_type}")
            plt_title = f"{app} - {test_source} - {test_type.capitalize()} Results"
            test_source_name = "Maestro Automated Test" if test_source == "automated" else "User Scroll Test"
            x_values = list(values.keys())
            y_values = list(values.values())
            plt.figure(figsize=(20,6))
            plt.bar(x_values, y_values, label=test_type.capitalize())
            plt.title(plt_title)
            plt.xlabel("Seconds")
            plt.ylabel("Measured")
            plt.legend()
            plt.savefig(f"./charts/{plt_title}.png")
            plt.clf()

    
```
